```python
fig, ax1 = plt.subplots(figsize=(12,4))

ax1.plot(results.avg_fitness, color='red')
ax1.fill_between(results.generation ,results.best_fitness, results.worst_fitness)
ax1.tick_params(axis='y')
plt.ylim()
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.plot(results.h, color='black')
ax2.plot([x for x in range(100000)], [0.002 for y in range(100000)], 'r--', color='red')
ax2.tick_params(axis='y')

for x in results.generation[results.new > 0]:
    plt.axvline(x, color='black', linewidth=0.5)
for x in results.generation[results.reset > 0]:
    plt.axvline(x, color='red', linewidth=0.5)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.xlim(0, 40000)
plt.show()
```

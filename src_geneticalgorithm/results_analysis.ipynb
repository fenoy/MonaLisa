```python
import pandas as pd
```

```python
crossover_results = pd.read_csv('../results/crossover/results.csv')
mutation_results = pd.read_csv('../results/mutation/results.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(crossover_results.generation*21, crossover_results.avg_fitness, color='red')
#plt.fill_between(crossover_results.generation*21 ,crossover_results.best_fitness, crossover_results.worst_fitness)
#for x in results.generation[results.new > 0]:
#    plt.axvline(x, color='black', linewidth=0.5)
plt.xlim(0,3000*21)
plt.ylim(0.9, 1)
plt.show()
```

```python
plt.plot(mutation_results.avg_fitness, color='red')
plt.fill_between(mutation_results.generation ,mutation_results.best_fitness, mutation_results.worst_fitness)
#for x in results.generation[results.new > 0]:
#    plt.axvline(x, color='black', linewidth=0.5)
plt.xlim(0, 61000)
plt.ylim(0.9, 1)
plt.show()
```

```python
plt.plot(mutation_results.generation/21/60, mutation_results.avg_fitness, color='blue', label='Hill Climbing')
plt.plot(crossover_results.generation/60, crossover_results.avg_fitness, color='red', label='Genetic Algorithm')
plt.fill_between(mutation_results.generation ,mutation_results.best_fitness, mutation_results.worst_fitness)
plt.xlim(0, 100)
plt.ylim(0.9, 1)
plt.xlabel('Execution time (min)')
plt.ylabel('Fitness')
plt.legend()
plt.show()
```
